```python
plt.figure(figsize=(6,4))
sns.histplot(df["Calories"], kde=True)
plt.title("Distribution of Calories Burned")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=df["Heart_Rate"], y=df["Calories"])
plt.title("Heart Rate vs Calories Burned")
plt.show()
```

```python
gender_counts = df["Gender"].value_counts()

labels = ["Males","Female"]
colors = ["#c3b091","#708191"] 

plt.title("Gender Distribution in Dataset")
plt.pie(gender_counts, labels=labels, colors=colors,autopct='%1.1f%%')
plt.show()
```

```python
bins = [0, 19, 29, 39, 49, 100]
labels = ["Teen (10-19)", "Young Adult (20-29)", "Adult (30-39)", 
          "Middle Age (40-49)", "Senior (50+)"]

df["Age_Group"] = pd.cut(df["Age"], bins=bins, labels=labels)
```

```python
df.head()
```

```python
age_group_counts = df["Age_Group"].value_counts().sort_index()
age_group_counts
```

```python
colors = ["#22505f","#8a9a5b","#6b4b44","#c3b091","#fff9e7"]

plt.title("Age Group Distribution")
plt.pie(age_group_counts,
        labels=age_group_counts.index, 
        colors=colors,
        autopct='%1.1f%%'
       )
plt.show()
```

```python

plt.figure(figsize=(7,5))
scatter = plt.scatter(
    df["Duration"],
    df["Calories"],
    c=df["Calories"],     # color based on calories
    cmap="magma"
)

plt.xlabel("Exercise Duration (minutes)")
plt.ylabel("Calories Burned")
plt.title("Duration vs Calories Burned")

plt.colorbar(scatter, label="Calories Burned")
plt.show()
```
